```python
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
df = pd.DataFrame(
    np.random.rand(100, 5),
    columns=['a', 'b', 'c', 'd', 'e']
)
```

```python
df
```

```python
profile = ProfileReport(df, title='Randomly sample data', html={'style':{'full_width':False}})
```

```python
profile.to_notebook_iframe()
```
